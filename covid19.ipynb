```python
kList=["Province/State","Country/Region","WHO Region","Date"]
nList=["Lat","Long","Confirmed","Deaths","Recovered","Active"]
```

```python
data=pd.get_dummies(data2,kList)
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=data.iloc[:,1:-1]
y = data.iloc[:,-1:].values
```

```python
x
```

```python
y
```

```python

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.33,random_state=0)
# Veri kümemizi test ve train şekinde bölüyoruz
```

```python

from sklearn.naive_bayes import GaussianNB # GaussianNB sınıfını import edildi
```

```python

gnb = GaussianNB() # GaussianNB sınıfından bir nesne üretildi
```

```python

gnb.fit(x_train, y_train.ravel()) # Makineyi eğitiliyor
```

```python

result = gnb.predict(x_test) # Test veri kümemizi veriyoruz ve x türü tahmin etmesini sağlıyoruz
```

```python

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,result)
print(cm)
# Karmaşıklık matrisi
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, result)
print(accuracy)
```
